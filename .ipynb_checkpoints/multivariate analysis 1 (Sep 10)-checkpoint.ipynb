```r
#3차원 그림
install.packages("mvtnorm")
library(mvtnorm)
```

```r
x.points = seq(-3, 3, length.out = 100)
y.points = x.points
z = matrix(0, 100, 100) #모든 값이 0 인 100 by 100 행렬 생성
```

```r
mu.vector = c(0,0) #평균벡터
sig.matrix = matrix(c(2,1,1,1), 2, 2) #분산공분산행렬
```

```r
for (i in 1:100) {
    for (j in 1:100) {
        z [i, j] = dmvnorm(c(x.points[i], y.points[j]), mean = mu.vector, sigma = sig.matrix)
    } # 이중 반복문을 돌며 z[i, j]에 (x, y)에 상응하는 값을 z 행렬에 저장 (높이를 저장하는거임)
}
```

```r
contour(x.points, y.points, z, main = "등고선 그래프")
```

```r
persp(x.points, y.points, z, phi = 45, theta = 45, main = "3차원 그래프") #phi: 상하 회전, theta: 좌우 회전 (각도)
```
