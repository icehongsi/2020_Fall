```r
library(ggplot2) #ggplot2는 tidyverse 안에 있음
library(tidyverse)
attach(iris)
head(iris)
```

```r
levels(iris$Species) # 범주형 변수 나열하기
```

```r
ggplot(data = iris, aes(x = Sepal.Length, y = Sepal.Width)) +
geom_point()
```

```r
myplot <- ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width))
myplot + geom_point() #점으로 표현
myplot + geom_line() #연속선으로 표현
```

```r
ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + geom_point(size = 3)
#Species에 따라 다른 색깔로 표현하기
#size 변수 설정에 따라 점의 크기를 조정할 수 있음
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) + geom_point(size = 3)
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) + geom_point(size = 3, aes(color = Species))
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) + geom_point(aes(color = Species, size = Sepal.Width))
#size는 종류에 따라 달라지게 하기
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) + geom_point(aes(color = Species, size = Sepal.Width), alpha = 0.2)
#alpha 매개변수는 점의 투명도를 설정
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) +
geom_boxplot() + geom_jitter() #boxplot 그리기 + 점도 겹쳐 그리기 (geom_jitter 이용)
```

```r
ggplot(iris, aes(x = Petal.Length, y = Petal.Width, color = Species)) +
geom_boxplot() + geom_jitter() + coord_flip() #>> X / Y축 반전
#boxplot 그리기 + 점도 겹쳐 그리기 (geom_jitter 이용)
```

```r
ggplot(iris, aes(Species, Petal.Width)) + geom_bar(stat = "identity", aes(fill = Species))
```

```r
long_iris <- gather(iris, key, value, 1:4)
head(long_iris)
head(iris)
```

```r
ggplot(long_iris, aes(x = Species, y = value)) + geom_bar(stat = "identity", aes(fill = key))
#꽃의 종류에 따라 여러 범주를 나타낸 그래프로 표현
```

```r
ggplot(long_iris, aes(x = Species, y = value)) +
geom_bar(stat = "identity", position = "fill", aes(fill = key))

#꽃의 종류에 따라 여러 범주를 나타낸 그래프로 표현 + 막대 총길이를 동일하게
```

```r
ggplot(iris, aes(Sepal.Length, Petal.Length)) +
geom_point() + facet_grid(~ Species) + stat_smooth(method = "lm", color = "red")
#facet_grid >> 종류에 따라 분할
#stat_smooth >> linear modeling
```

```r
ggplot(iris, aes(Sepal.Length, Petal.Length)) +
geom_point(aes(color = Species)) + facet_grid(~ Species) + stat_smooth(method = "lm", color = "red") +
theme(panel.grid.major = element_line(color = "black"))
#facet_grid >> 종류에 따라 분할
#stat_smooth >> linear modeling
```

```r
ggplot(long_iris, aes(x = factor(1), y = value)) + geom_bar(stat = "identity", position = "fill", aes(fill = key)) +
facet_grid(~ Species)
```

```r
ggplot(long_iris, aes(x = factor(1), y = value)) + geom_bar(stat = "identity", position = "fill", aes(fill = key)) +
facet_grid(~ Species) + coord_polar(theta = "y")
```
